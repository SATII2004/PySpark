###Schema Definition


```python
df.printSchema()
```

###SELECT

```python
from pyspark.sql.functions import col
df.select(col('Name'),col('Abbreviation'),col('Numeric')).display()
```

###ALIAS

```python
df.select(col('Abbreviation').alias('Numeric')).display()
```

```python
df.display()
```

### FILTER

```python
df.filter(col('Abbreviation')=='Aug.').display()
```

```python
df.filter((col('Abbreviation')=='Aug.') & (col('Numeric')==8) ).display()
```

###Eith Column Renamed

```python
df.withColumnRenamed('Numeric-2','Number').display()
```

###withColumn

```python
from pyspark.sql.functions import lit

df=df.withColumn('Flag',lit("new"))
df.display()
```

```python
from pyspark.sql.functions import col

df = df.withColumn(
    'multiply',
    col('Numeric').cast('double') * col('Numeric-2').cast('double')
)
display(df)
```

```python
from pyspark.sql.functions import regexp_replace
df.withColumn('Abbreviation',regexp_replace(col('Abbreviation'),"May","Mahi")).display()
```

###SORT

```python
df.sort(col('Numeric').desc()).display()
```
