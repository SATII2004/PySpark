```python
df=spark.read.option("header",True).option("inferSchema",True).csv("/Volumes/workspace/satish/files/airlines_flights_data.csv")

display(df)
```

```python
df.columns
df.count()
```

```python
df.drop("index")
df.show(5)
```

```python
df.select("airline","source_city").show(4)
```

```python
df.filter(df.source_city == "Delhi").show(10)
```

```python
from pyspark.sql.functions import when

df=df.withColumn(
    "duration_category",when(df.duration >5,"Long Flight").otherwise("Short Flight")
)

df.select("duration","duration_category").show(10)
```

```python
df.groupBy("airline").count().orderBy("count",ascending=False).show()
```

```python
from pyspark.sql.functions import avg

df.groupBy("airline").agg(avg("price").alias("avg_price")).show()
```

```python
df.orderBy(df.price.desc()).show()
```

```python
df.groupBy("source_city","destination_city").agg(avg("price").alias("avg_price")).orderBy("avg_price",ascending=False).show(5)
```

```python
from pyspark.sql.functions import col

df_filtered = df.filter(
    col("stops").try_cast("int") == 2
)
df_filtered.count()
```
