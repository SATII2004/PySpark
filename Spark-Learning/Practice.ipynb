```python
from pyspark.sql.functions import when

df=df.withColumn(
    "duration_category",when(df.duration >5,"Long Flight").otherwise("Short Flight")
)

df.select("duration","duration_category").show(10)
```

```python
df.groupBy("airline").count().orderBy("count",ascending=False).show()
```

```python
from pyspark.sql.functions import avg

df.groupBy("airline").agg(avg("price").alias("avg_price")).show()
```

```python
df.groupBy("source_city","destination_city").agg(avg("price").alias("avg_price")).orderBy("avg_price",ascending=False).show(5)
```

```python
from pyspark.sql.types import StructType, StructField,StringType,IntegerType,FloatType

schema=StructType([
    StructField("airline",StringType(),True),
    StructField("flight",StringType(),True),
    StructField("source_city",StringType(),True),
    StructField("destination_city",StringType(),True),
    StructField("price",StringType(),True),

])

df_schema=spark.read.schema(schema).csv("/Volumes/workspace/satish/files/airlines_flights_data.csv")

df_schema.printSchema()

```
